### Rolling Statistics-Moving Average

```python
df1=df.rename(columns={"#Passengers":"Passengers"})
```

```python
df1['Moving Average']=df1.Passengers.rolling(12).mean()
```

```python
df1["MA2"]=df1.Passengers.rolling(12).std()
```

```python
df1
```

```python
df1["Passengers"].plot(grid=True)
df1["Moving Average"].plot(grid=True)
df1["MA2"].plot(grid=True)
```

From here it can be observed that moving mean has a positive slope which means the data has an upward trend. The standard deviation is relatively gradual in increasing. A period of 12 is chosen for plotting rolling statistics because the data depicts seasonal variation in every 12 months.

```python
import statsmodels
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller(df1["Passengers"])
```

The test statistic is greater than all 3 critical values. This clearly indicates non stationarity.

```python
df1['log']=np.log(df1['Passengers'])
```

Log of values is taken to stabilize the variance of the series.

```python
df1["1st difference"]=df1["log"].diff()
```

Since from above conclusions it is evident that the data has inherent upward trend, we perform differencing and calculate first difference.

```python
df2=df1.dropna(axis=0)
```

```python
df2.head()
```

```python
df2.shape
```

```python
adfuller(df2["1st difference"])
```

The test statistic (-3.122) is greater than the critical values at 1%. Hence, stationarity has not yet been achieved.

As, the data depicts seasonal behaviour as well, we perform second differencing too and try to achieve stationarity.

```python
df2["2nd difference"]=df2["1st difference"].diff()
```

```python
df2.head()
```

```python
df3=df2.dropna(axis=0)
```

```python
adfuller(df3["2nd difference"])
```

The test statistic is less than critical values for all significance levels. Hence , stationarity has been achieved.

```python
df2["1st difference"].plot()
```

The upward trend is seen to have been removed. But seasonal variations are still visible

```python
df3["2nd difference"].plot()
```

Now, the seasonal variations have also been removed.

### ACF AND PACF Plots 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_acf(df3["2nd difference"])
plt.show()
```

```python
plot_pacf(df3["2nd difference"])
plt.show()
```

Finding the order of ARMA model by iterating 

```python
from statsmodels.tsa.arima_model import ARMA
import itertools
```

```python
p=q=range(0,8)
pq=list(itertools.product(p,q))
list1=[]
list2=[]
for order in pq:
    try:
        model=ARMA(df3["2nd difference"],order=order)
        model_fit=model.fit()
        list1.append(order)
        list2.append(model_fit.aic)
    except: 
        continue
```

```python
dlist1=pd.DataFrame(list1)
dlist1.rename(columns={0:"p",1:"q"},inplace=True)
dlist2=pd.DataFrame(list2)
dlist2.rename(columns={0:"Aic"},inplace=True)
table=pd.concat([dlist1,dlist2],axis=1)
```

```python
table
```

```python
table["Aic"].min()
```

Finding the order corresponding to minimum AIC

```python
data=table.query("Aic==Aic.min()")
```

```python
data
```

Hence order is (7,4)

Fitting the model

```python
model2=ARMA(df3["2nd difference"],order=(7,4))
```

```python
model2_fit=model2.fit()
```

```python
pred=model2_fit.predict(start='1950-01-01',end='1960-12-01')
```

Plotting the predicted values  against the actual values

```python
df2["2nd difference"].plot()
pred.plot()
```

The fit is good.

Scaling back the data

```python
pr=model2_fit.predict()
```

```python
pr1=pr.to_frame()
```

```python
pr2= pr1[0]+df2["1st difference"].shift(1)
```

```python
pr3= pr2+ df1["log"].shift(1)
```

```python
pr3
```

```python
pr4=np.exp(pr3)
```

Plotting Actual vs Predicted data (ijn original form)

### Model Error
